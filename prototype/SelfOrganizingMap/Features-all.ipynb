```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import preprocessing
import sklearn.metrics as metrics
import math
from scipy.stats import norm, skew
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
label_name = "math"
embedding_type = "perf" # time or perf

train = pd.get_dummies(pd.read_csv(f'../../dataset/{embedding_type}/{label_name}_dataset.csv'))
```

```python
corr = train.corr()
highest_corr_features = corr.index[abs(corr["label"])>0.21]
plt.figure(figsize=(15,15))
g = sns.heatmap(train[highest_corr_features].corr(),annot=True)
```
